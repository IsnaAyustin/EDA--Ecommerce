```python
def check_plot(df, variable):
    plt.figure(figsize=(16, 4))

    # histogram
    plt.subplot(1, 3, 1)
    sns.histplot(df[variable], bins=30)
    plt.title('Histogram')

    # plot Q-Q
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.ylabel('Variable quantiles')

    # box plot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')

    plt.show()
```

terdapat Outlier dari ketiga kolom tersebut maka akan dilakukan Handling Oulier

```python
def find_outlier_boundary(data_ecommerce, variable):

    # Mari kita hitung batas luar yang merupakan outlier

    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

    lower_boundary = df[variable].quantile(0.25) - (IQR * 1.5)
    upper_boundary = df[variable].quantile(0.75) + (IQR * 1.5)

    return upper_boundary, lower_boundary
```

```python
# batas atas dan batas bawah untuk Quantity
full_occup_upper_limit, full_occup_lower_limit = find_outlier_boundary(data_ecommerce, 'Quantity')
full_occup_upper_limit, full_occup_lower_limit
```

```python
# batas atas dan batas bawah untuk UnitPrice
full_occup_upper_limit, full_occup_lower_limit = find_outlier_boundary(data_ecommerce, 'UnitPrice')
full_occup_upper_limit, full_occup_lower_limit
```

```python
# Trimming outliers from the training data
data_ecommerce = data_ecommerce[(data_ecommerce['Quantity'] <= full_occup_upper_limit) & (data_ecommerce['Quantity'] >= full_occup_lower_limit)]
```

```python
# Trimming outliers from the training data
data_ecommerce = data_ecommerce[(data_ecommerce['UnitPrice'] <= full_occup_upper_limit) & (data_ecommerce['UnitPrice'] >= full_occup_lower_limit)]
```

3. 5 produk paling laris di negara united kingdom?

```python
# Filter data UK dan sort by Quantity
uk_sales = df[df['Country'] == 'United Kingdom']
top_5_products = uk_sales.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(5)

# Tambah informasi revenue
product_details = uk_sales.groupby('Description').agg({'Quantity':'sum', 'Revenue':'sum'})
top_5_details = product_details.loc[top_5_products.index]
```

```python
plt.figure(figsize=(10,6))
bars = sns.barplot(
    x=top_5_products.values,
    y=top_5_products.index,
    palette="viridis",
    hue=top_5_products.index,
    legend=False
)

plt.title('TOP 5 Produk Terlaris di UK (Berdasarkan Quantity)', fontsize=16, pad=20)
plt.xlabel('Total Quantity Terjual', fontsize=12)
plt.ylabel('Produk', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.3)

# Tambah label nilai
for i, value in enumerate(top_5_products.values):
    bars.text(value+5, i, f"{value} unit", va='center', fontsize=11)

plt.tight_layout()
plt.show()
```

dari grafik barplot diatas produk yang paling mendominasi untuk negara UK adalah Jumbo Bag Red Retrospot dengan terjual 310 unit. terlihat dari grafik diatas 5  barang yang paling mendominasi penjualan di UK adalah produk alat rumah tangga.

# Date Time

4. Berapa total nilai Revenue perbulan?

```python
# Mengubah kolom 'Invoice Date' menjadi tipe DateTime
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"],format="%m/%d/%Y %H:%M", errors="coerce")
```

```python
df['Month'] = df['InvoiceDate'].dt.to_period('M')
monthly_Revenue = df.groupby('Month')['Revenue'].sum().reset_index()
monthly_Revenue
```

```python
plt.figure(figsize=(12, 8))
monthly_Revenue.plot(kind='line', marker='o', color='red', linestyle='-')

for i, total in enumerate(monthly_Revenue["Revenue"]):
    plt.text(i, total + (0.001 * (monthly_Revenue["Revenue"]).max()), f"{total:,.2f}",
             ha="center", fontsize=9, color="black")

plt.title('Tren Pendapatan Bulanan', fontsize=14)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Total Pendapatan', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

dari grafik Total nilai Revenue diatas terlihat bahwa nilai tertinggi ada pada bulan ke-11 dengan nilai 17,948.84 dan nilai terendah ada pada bulan ke-2 dengan nilai 6152.46.

# Crosstab

5. Bagaimana distribusi penjualan berdasarkan negara?

```python
# Crosstab dengan aggregasi revenue
revenue_by_country = pd.crosstab(
    index=df['Country'],
    columns='Total Revenue',
    values=df['Revenue'],
    aggfunc='sum'
).sort_values('Total Revenue', ascending=False).reset_index()

revenue_by_country
```

```python
# Visualisasi
plt.figure(figsize=(10,5))
sns.barplot(
    x=revenue_by_country.index,
    y=revenue_by_country['Total Revenue'],
    palette='viridis',
    legend=False
)


plt.title('Distribusi Revenue per Negara')
plt.xlabel('Negara')
plt.ylabel('Total Revenue ($)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

distribusi nilai tertinggi Revenue ada di negara UK (United Kingdom) dengan nilai 93011.36, dan diikuti dengan negara lain sperti Netherlands, Germany dll, jarak nilai UK dengan negara lain sperti Netherlands sangat jauh yaitu sekitar 57. ini menunjukan bahwa UK adalah negara dengan total Revenue paling tinggi

## 3. Lakukan analisa multivariate pada kolom numerikal berupa :
- Correlation Heatmap + insight / observation **(10 poin)**

```python
df.head()
```

```python
# Pilih hanya kolom numerik
numeric_df = df.select_dtypes(include=[np.number])
```

```python
# Hitung korelasi
corr_matrix = numeric_df.corr()
```

```python
# Visualisasi heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    center=0,
    fmt='.2f',
    linewidths=0.5,
    annot_kws={'size': 10}
)
plt.title('Correlation Heatmap Kolom Numerik', pad=20)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

hubungan antara Quantity dengan Revenue memiliki hubungan korelasi (0.31). Peningkatan quantity berhubungan dengan peningkatan revenue.

hubungan antara Quantity dengan UnitPrice (-0,37). Produk dengan harga lebih rendah cenderung dibeli dalam quantity lebih besar.
